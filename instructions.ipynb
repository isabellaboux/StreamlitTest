## 💡 Useful Streamlit-Commands & Features to create your own little project

- **`st.set_page_config`** – Sets the title, layout, and icon of the app (called early in the script).  
- **`st.tabs`** – Creates tabs for clear subdivisions within a page.  
- **`st.columns`** – Splits the screen into columns for parallel content (e.g., KPIs).  
- **`st.metric`** – Displays key figures with an optional change indicator.  
- **`st.cache_data`** – Caches return values of functions with (immutable) data to save loading time.  
- **`st.cache_resource`** – Caches “heavy” objects like models or database connections.  
- **`st.form`** – Groups multiple inputs into a form that is only processed after a button click.  
- **`st.form_submit_button`** – Button to submit an `st.form` and trigger processing.  
- **`st.status`** – Displays a status block with progress, e.g., during training or loading.  
- **`st.toast`** – Small pop-up notification at the bottom right for brief feedback.  
- **`st.data_editor`** – Editable table view with column configuration and inline changes.  
- **`st.download_button`** – Offers files (CSV, Excel, images, etc.) for download.  
- **`st.plotly_chart`** / **`st.pydeck_chart`** – Embed interactive visualizations with Plotly or maps with Pydeck.  
- **`st.Page`** – Defines a page for multipage apps (modern approach).  
- **`st.navigation`** – Creates navigation between multiple `st.Page` objects (navbar or sidebar).  
- **`st.page_link`** – Creates links to other app pages or external URLs, e.g., for custom navigation.  
- **`st.chat_message`** / **`st.chat_input`** – Builds simple chat UIs with message layout and input field.  
- **`st.session_state`** – Stores states (variables) that persist between interactions.  
- **`st.connection`** – Establishes a connection to a database or external data source.  
- **`st.secrets`** – Manages sensitive information (e.g., API keys) securely in the app.  
- **`st.logo`** – Adds a logo to the navigation (only with `st.navigation`).
